```python
### Basic libraries
# !pip install pandas
import pandas as pd

### Visualization
# !pip install matplotlib
import matplotlib.pyplot as plt

### Association rules
# !pip install mlxtend
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder

### Functions
from utils3 import association_rules_pipeline

### Path
path = 'G:\\O meu disco\\Universidade\\2nd year\\2nd semester\\Machine Learning II\\Project\\'
```

```python
# Load customer basket dataset
customer_basket = pd.read_csv(path + 'Customer Basket Dataset.csv')

# Load customer segmentation datasets
gamers = pd.read_csv(path + 'gamers.csv', index_col = 0)
spenders = pd.read_csv(path + 'spenders.csv', index_col = 0)
veggies = pd.read_csv(path + 'veggies.csv', index_col = 0)
promo_hunters = pd.read_csv(path + 'promo_hunters.csv', index_col = 0)
alcoholics = pd.read_csv(path + 'alcoholics.csv', index_col = 0)
parents = pd.read_csv(path + 'parents.csv', index_col = 0)
young_adults = pd.read_csv(path + 'young_adults.csv', index_col = 0)
low_standards = pd.read_csv(path + 'low_standards.csv', index_col = 0)
olds = pd.read_csv(path + 'olds.csv', index_col = 0)
supermarkets = pd.read_csv(path + 'supermarkets.csv', index_col = 0)
```

```python
# Create a dictionary with the clusters
clusters = {'gamers': gamers, 
            'spenders': spenders, 
            'veggies': veggies, 
            'promo_hunters': promo_hunters, 
            'alcoholics': alcoholics, 
            'parents': parents, 
            'young_adults': young_adults, 
            'low_standards': low_standards, 
            'olds': olds, 
            'supermarkets': supermarkets
            }

# Display the association rules for each cluster
# The association rules are sorted by lift in descending order
# Only the top 60 association rules are displayed
for cluster_name, cluster in clusters.items():
    print(cluster_name.upper())
    display(association_rules_pipeline(cluster, customer_basket).sort_values('lift', ascending = False)[:60])
```
